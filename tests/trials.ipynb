```python
from llama_cpp import Llama 
import json 
# from ..krishi_mitra.router import IntentRouter
```

```python
model_path = "../models/Qwen3-8B-Q6_K.gguf"
```

```python
#Loading Qwen 
llm = Llama(
    model_path=model_path,
    n_gpu_layes=-1,
    n_ctx=8192,
    verbose=False,
    n_batch = 1024,
    falsh_attn=True

)
```

```python
output = llm(
      "Question: What district is paithan in?  Answer: ", # Prompt
      max_tokens=50, # Generate up to 32 tokens, set to None to generate up to the end of the context window
      stop=["Q:", "\n"], # Stop generating just before the model would generate a new question
      echo=True # Echo the prompt back in the output
) # Generate a completion, can also call create_completion
output
```

```python
router = IntentRouter(llm)
```

```python
queries = [
    "Plan a crop rotation for cotton and soybean.",
    "What is the best sowing time for Wheat?",
    "My tomato leaves have black spots.",
    "Current price of Onion in Lasalgaon?",
    "How do I apply for a tractor loan?",
    "Hello, good morning."
]

print("--- ROUTER TEST ---")
for q in queries:
    decision = router.classify(q, debug=True)
    print(f"Query: '{q}'\n -> Agent: {decision.get('agent')} | Think: {decision.get('think')}\n")
```
